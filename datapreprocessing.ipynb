```python
# Convert string representations of lists into actual lists
d1['Amenities'] = d1['Amenities'].apply(literal_eval)
```

```python
# Group by city and concatenate the amenities into unique lists
def concat_unique(amenity_lists):
    unique_amenities = set()
    for amenity_list in amenity_lists:
        unique_amenities.update(amenity_list)
    return list(unique_amenities)

# Group by city and apply the custom aggregation function
grouped_data = d1.groupby('City')['Amenities'].agg(concat_unique).reset_index()

# Rename the column for clarity
grouped_data.rename(columns={'Amenities': 'Combined_Amenities'}, inplace=True)
```
